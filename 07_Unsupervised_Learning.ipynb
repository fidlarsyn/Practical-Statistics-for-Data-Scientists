# **Principal Components Analysis (PCA)**

**Kumpulan Kode:**

```python
from sklearn.decomposition import PCA
import pandas as pd

# Menjalankan PCA (misal pada data return saham)
pca = PCA(n_components=5)
pca.fit(oil_px)

# Mendapatkan weights (loadings) untuk tiap komponen
loadings = pd.DataFrame(pca.components_.T, index=oil_px.columns)

# Melihat persentase variansi yang dijelaskan oleh tiap komponen
print(pca.explained_variance_ratio_)
```

# **Hierarchical Clustering**
Berbeda dengan K-Means, metode ini memulai dengan setiap record sebagai cluster-nya sendiri, lalu secara bertahap menggabungkan cluster terdekat hingga semua tergabung. Hasilnya divisualisasikan dalam bentuk Dendrogram (diagram pohon) yang memungkinkan kita melihat struktur hirarki tanpa harus menentukan jumlah cluster di awal.


**Kumpulan Kode:**

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Melakukan pengklasteran hirarkis menggunakan metode 'complete'
Z = linkage(df, method='complete')

# Visualisasi Dendrogram
fig, ax = plt.subplots(figsize=(5, 5))
dendrogram(Z, labels=df.index, ax=ax)
plt.show()
```

# **Model-Based Clustering**
Teknik ini mengasumsikan bahwa data berasal dari campuran beberapa distribusi probabilitas (biasanya distribusi normal multivariate). Metode ini lebih ketat secara statistik dan menggunakan metrik BIC (Bayesian Information Criteria) untuk menentukan jumlah cluster yang paling optimal secara otomatis.

**Kumpulan Kode:**

```python
from sklearn.mixture import GaussianMixture

# Melatih model campuran Gaussian (Gaussian Mixture Model)
mclust = GaussianMixture(n_components=2).fit(df)

# Prediksi label cluster
df['cluster'] = mclust.predict(df)

# Melihat nilai BIC (semakin rendah biasanya semakin baik)
print(mclust.bic(df))
```
