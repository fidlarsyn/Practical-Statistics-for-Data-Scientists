# **Naive Bayes**

**Kumpulan Kode:**

```python
from sklearn.naive_bayes import MultinomialNB

# Menyiapkan fitur (X) dan target (y)
# Fitur kategorikal harus dikonversi ke variabel dummy terlebih dahulu
X = pd.get_dummies(loan_data[predictors])
y = loan_data[outcome]

# Melatih model
naive_model = MultinomialNB()
naive_model.fit(X, y)

# Prediksi probabilitas untuk data baru
# Mengembalikan probabilitas [Paid off, Default]
probabilities = naive_model.predict_proba(new_loan)
```

# **Regresi Logistik (Logistic Regression)**
Mirip dengan regresi linear berganda, namun digunakan untuk hasil biner. Model ini memetakan probabilitas hasil (0 hingga 1) menggunakan fungsi logit (log-odds). Teknik ini sangat populer karena kecepatan komputasinya dan kemudahan interpretasi melalui nilai odds ratio.

**Kumpulan Kode:**

```python
from sklearn.linear_model import LogisticRegression

# Melatih model logistik (C besar berarti tanpa regularisasi)
logit_reg = LogisticRegression(penalty='l2', C=1e42, solver='liblinear')
logit_reg.fit(X, y)

# Interpretasi koefisien menjadi Odds Ratio
import numpy as np
print(np.exp(logit_reg.coef_))
```
