# **Tree Models (Model Pohon)**
Model pohon menggunakan algoritme partisi rekursif untuk membagi data menjadi sub-bagian yang semakin murni (homogeneous) berdasarkan aturan "jika-maka". Kualitas pemisahan diukur menggunakan metrik Gini Impurity atau Entropy, dan pohon harus dikontrol (misal melalui grid search parameter kedalaman) agar tidak terjadi overfitting.

**Kumpulan Kode:**

```python
from sklearn.tree import DecisionTreeClassifier

# Melatih model pohon keputusan
# max_depth dan min_samples_split membantu mencegah overfitting
model = DecisionTreeClassifier(max_depth=5, min_samples_split=20)
model.fit(X, y)

# Prediksi label kelas
y_pred = model.predict(X)
```

# **Bagging and the Random Forest**
Bagging (bootstrap aggregating) melatih banyak model secara paralel pada sampel data yang diambil secara acak dengan pengembalian untuk mengurangi varians. Random Forest adalah varian bagging yang sangat kuat karena juga memilih subset variabel secara acak pada setiap pemisahan pohon, yang memungkinkan model mengukur tingkat kepentingan setiap fitur (variable importance).

**Kumpulan Kode:**

```python
from sklearn.ensemble import RandomForestClassifier

# Melatih Random Forest dengan 500 pohon
rf = RandomForestClassifier(n_estimators=500, oob_score=True)
rf.fit(X, y)

# Mengambil Variable Importance (Gini Decrease)
importances = rf.feature_importances_

# Mengecek Out-of-Bag (OOB) score sebagai estimasi akurasi
print(f'OOB Score: {rf.oob_score_}')
```
