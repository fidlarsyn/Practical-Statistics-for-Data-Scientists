# **Estimasi Lokasi (Estimates of Location)**
Menentukan nilai yang "tipikal" atau pusat dari data (tendensi sentral). Metrik utama meliputi Mean (rata-rata), Median (nilai tengah yang lebih robust terhadap outlier), dan Trimmed Mean (rata-rata setelah membuang nilai ekstrem).

**Kumpulan Kode:**

```python
import numpy as np
from scipy.stats import trim_mean

# Mean dan Median
print(state['Population'].mean())
print(state['Population'].median())

# Trimmed Mean (membuang 10% data di tiap ujung)
print(trim_mean(state['Population'], 0.1))

# Weighted Mean (Rata-rata berbobot populasi)
print(np.average(state['Murder.Rate'], weights=state['Population']))
```

# **Estimasi Variabilitas (Estimates of Variability)**
Mengukur seberapa tersebar atau terdifusinya data di sekitar nilai pusat. Metrik populer adalah Standar Deviasi, Varians, dan Interquartile Range (IQR). Metrik yang robust seperti Median Absolute Deviation (MAD) disarankan jika data mengandung banyak outlier.

**Kumpulan Kode:**

```python
from statsmodels import robust

# Standar Deviasi
print(state['Population'].std())

# Interquartile Range (IQR = Kuartil 75% - Kuartil 25%)
print(state['Population'].quantile(0.75) - state['Population'].quantile(0.25))

# Robust Median Absolute Deviation (MAD)
print(robust.scale.mad(state['Population']))
```

# **Menjelajahi Distribusi Data (Exploring the Data Distribution)**
EDA menggunakan visualisasi untuk melihat bagaimana data tersebar. Boxplot efektif untuk melihat persentil dan outlier, Histogram menampilkan frekuensi nilai dalam bin (kotak), dan Density Plot memberikan garis kontinu yang halus untuk melihat bentuk distribusi.

**Kumpulan Kode:**

```python
import matplotlib.pyplot as plt

# Boxplot untuk populasi dalam jutaan
ax = (state['Population']/1_000_000).plot.box()
plt.show()

# Histogram
ax = (state['Population'] / 1_000_000).plot.hist(bins=10)
plt.show()

# Density Plot
ax = state['Murder.Rate'].plot.density()
plt.show()
```
