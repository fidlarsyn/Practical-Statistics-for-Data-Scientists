# **Pengambilan Sampel Acak dan Bias Sampel (Random Sampling and Sample Bias)**
Sampel adalah subset dari dataset yang lebih besar (populasi). Sampling Acak memastikan setiap anggota populasi memiliki peluang yang sama untuk terpilih guna mencapai keterwakilan (representativeness). Bias Sampel terjadi jika sampel berbeda secara sistematis dan non-acak dari populasi yang diwakilinya.

# **Bias Seleksi (Selection Bias)**
Bias yang dihasilkan dari cara pengamatan dipilih, baik secara sadar maupun tidak. Ini mencakup Data Snooping (mencari pola hingga menemukan sesuatu yang menarik namun fana) dan Regresi ke Rata-rata (Regression to the Mean), di mana nilai ekstrem cenderung diikuti oleh nilai yang lebih mendekati pusat pada pengukuran berikutnya.

# **Distribusi Sampling dari Suatu Statistik (Sampling Distribution of a Statistic)**
Penggambarkan bagaimana statistik sampel (seperti rata-rata) bervariasi jika kita mengambil banyak sampel dari populasi yang sama. Teorema Limit Pusat menyatakan bahwa distribusi rata-rata sampel akan menyerupai kurva normal (lonceng) meskipun populasi asalnya tidak terdistribusi normal, asalkan ukuran sampel cukup besar.

# **Bootstrap**
Teknik praktis untuk mengestimasi distribusi sampling dengan mengambil sampel tambahan secara berulang dengan pengembalian (with replacement) dari sampel asli itu sendiri. Bootstrap sangat kuat untuk menghitung kesalahan standar dan interval kepercayaan tanpa asumsi matematika yang rumit.

**Kumpulan Kode:**

```python
import pandas as pd
from sklearn.utils import resample

# Implementasi Bootstrap untuk mencari distribusi Median
results = []
for nrepeat in range(1000):
    sample = resample(loans_income) # Mengambil sampel dengan pengembalian
    results.append(sample.median())

results = pd.Series(results)
print('Statistik Bootstrap:')
print(f'Original Median: {loans_income.median()}')
print(f'Bias: {results.mean() - loans_income.median()}')
print(f'Standard Error: {results.std()}')
```

# **Interval Kepercayaan (Confidence Intervals)**
Cara standar untuk menyajikan estimasi statistik sebagai rentang nilai, bukan angka tunggal. Interval ini mencerminkan tingkat ketidakpastian yang terkait dengan estimasi tersebut. Bootstrap adalah metode yang efektif untuk membangun interval ini.

# **Distribusi Ekor Panjang (Long-Tailed Distributions)**
Data dunia nyata sering kali tidak normal dan memiliki "ekor panjang," di mana nilai ekstrem jauh lebih sering terjadi daripada yang diprediksi oleh distribusi normal (sering disebut sebagai fenomena Black Swan).

# **Distribusi-t Student (Student’s t-Distribution)**
Mirip dengan distribusi normal tetapi dengan ekor yang lebih tebal. Distribusi ini digunakan sebagai referensi untuk distribusi rata-rata sampel, terutama ketika ukuran sampel kecil.

# **Distribusi Binomial (Binomial Distribution)**
Digunakan untuk memodelkan hasil biner (ya/tidak, sukses/gagal). Inti dari distribusi ini adalah serangkaian percobaan di mana setiap percobaan memiliki peluang sukses tertentu.
