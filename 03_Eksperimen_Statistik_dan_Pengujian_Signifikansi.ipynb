# **A/B Testing**
Eksperimen dengan dua kelompok (A dan B) untuk menentukan perlakuan (treatment) mana yang lebih unggul. Biasanya membandingkan perlakuan baru dengan kelompok kontrol (standar atau tanpa perlakuan). Teknik ini sangat populer dalam desain web dan pemasaran untuk menguji efektivitas elemen visual atau harga.

# **Hipotesis Tests (Uji Hipotesis)**
Bertujuan untuk menentukan apakah peluang acak bertanggung jawab atas efek yang diamati.
* **Hipotesis Nol (H0)**: Asumsi bahwa efek yang diamati hanyalah produk dari kebetulan.
* **Hipotesis Alternatif (Ha)**: Kebalikan dari hipotesis nol, yaitu apa yang ingin kita buktikan.

# **Resampling (Uji Permutasi)**
Prosedur untuk menguji hipotesis dengan menggabungkan hasil dari beberapa grup dan membaginya kembali secara berulang untuk melihat seberapa sering perbedaan ekstrem terjadi secara tidak sengaja.

**Kumpulan Kode (Logika Uji Permutasi):**

```python
import numpy as np
import pandas as pd

# Fungsi permutasi sederhana untuk membandingkan rata-rata dua grup
def perm_fun(x, nA, nB):
    n = nA + nB
    idx_B = np.random.choice(range(n), nB, replace=False)
    idx_A = list(set(range(n)) - set(idx_B))
    return x.loc[idx_B].mean() - x.loc[idx_A].mean()

# Contoh menghitung proporsi di mana permutasi melebihi perbedaan observasi (p-value)
# np.mean(perm_diffs > mean_b - mean_a)
```
